```python
%%writefile UnusualAnimation.py
from manim import *

class UnusualAnimation(Scene):
    def construct(self):
        # Ваш код анимации здесь (см. выше)
        squares = VGroup()
        colors = [RED, ORANGE, YELLOW, GREEN, BLUE, PURPLE]
        for i in range(6):
            square = Square(side_length=1, color=colors[i % len(colors)])
            square.shift(RIGHT * i * 1.2)
            squares.add(square)

        self.play(
            LaggedStart(
                *[square.animate.scale(2) for square in squares],
                lag_ratio=0.2
            ),
            run_time=2
        )
        self.wait(1)

        self.play(
            LaggedStart(
                *[Rotate(square, angle=PI) for square in squares],
                lag_ratio=0.2
            ),
            LaggedStart(
                *[square.animate.set_color(colors[(i+1)%len(colors)]) for i, square in enumerate(squares)],
                lag_ratio=0.2
            ),
            run_time=2
        )
        self.wait(1)

        text = Text("1+1=11", font_size=48)
        self.play(Write(text), run_time=2)
        self.wait(1)

        self.play(
            text.animate.scale(1.2),
            rate_func=there_and_back,
            run_time=1
        )
        self.wait(1)
```

```python
%%writefile neural_network_visualization.py
from manim import *

class NeuralNetworkVisualization(Scene):
    def construct(self):
        # Заголовок
        title = Text("Визуализация Нейронной Сети", font_size=48)
        self.play(Write(title), run_time=2)
        self.wait(1)
        self.play(title.animate.to_edge(UP), run_time=1)

        # Построение графика функции
        axes = Axes(
            x_range=[-3, 3, 1],
            y_range=[-1.5, 1.5, 0.5],
            x_length=7,
            y_length=4,
            axis_config={"include_numbers": True},
        )
        function = axes.plot(lambda x: np.sin(x), color=BLUE)
        function_label = axes.get_graph_label(function, label="f(x) = \\sin(x)")

        self.play(Create(axes), run_time=2)
        self.play(Create(function), Write(function_label), run_time=2)
        self.wait(2)
        self.play(FadeOut(axes), FadeOut(function), FadeOut(function_label), run_time=1)

        # Визуализация нейронной сети
        nn = self.create_neural_network()
        self.play(FadeIn(nn), run_time=2)
        self.wait(1)
        self.animate_neural_network(nn)
        self.wait(2)
        self.play(FadeOut(nn), run_time=2)

        # Завершающий текст
        closing_text = Text("Спасибо за внимание!", font_size=48)
        self.play(Write(closing_text), run_time=2)
        self.wait(2)

    def create_neural_network(self):
        # Определение слоев (число нейронов в каждом слое)
        input_layer = self.create_layer(3)
        hidden_layer = self.create_layer(4)
        output_layer = self.create_layer(2)

        # Позиционирование слоев
        input_layer.shift(LEFT * 4)
        hidden_layer.shift(ORIGIN)
        output_layer.shift(RIGHT * 4)

        # Соединение слоев
        connections = self.connect_layers(input_layer, hidden_layer)
        connections += self.connect_layers(hidden_layer, output_layer)

        # Группировка всех элементов
        nn_group = VGroup(input_layer, hidden_layer, output_layer, *connections)

        return nn_group

    def create_layer(self, num_neurons):
        neurons = VGroup()
        for i in range(num_neurons):
            neuron = Circle(radius=0.3, color=BLUE, fill_opacity=0.5)
            neuron.shift(DOWN * (i - (num_neurons - 1) / 2) * 1.2)
            neurons.add(neuron)
        return neurons

    def connect_layers(self, layer1, layer2):
        connections = VGroup()
        for neuron1 in layer1:
            for neuron2 in layer2:
                line = Line(neuron1.get_right(), neuron2.get_left(), stroke_width=2)
                connections.add(line)
        return connections

    def animate_neural_network(self, nn_group):
        # Анимация прохождения данных через сеть
        input_layer = nn_group[0]
        hidden_layer = nn_group[1]
        output_layer = nn_group[2]
        connections = nn_group[3:]

        # Выделение входных нейронов
        self.play(
            *[neuron.animate.set_fill(YELLOW) for neuron in input_layer],
            run_time=1
        )
        self.wait(0.5)

        # Передача активации в скрытый слой
        self.play(
            *[ShowPassingFlash(line.copy().set_color(YELLOW), run_time=1) for line in connections[:len(connections)//2]],
            *[neuron.animate.set_fill(ORANGE) for neuron in hidden_layer],
            run_time=2
        )
        self.wait(0.5)

        # Передача активации в выходной слой
        self.play(
            *[ShowPassingFlash(line.copy().set_color(ORANGE), run_time=1) for line in connections[len(connections)//2:]],
            *[neuron.animate.set_fill(RED) for neuron in output_layer],
            run_time=2
        )
        self.wait(1)

        # Сброс цветов
        self.play(
            *[neuron.animate.set_fill(BLUE, opacity=0.5) for neuron in input_layer],
            *[neuron.animate.set_fill(BLUE, opacity=0.5) for neuron in hidden_layer],
            *[neuron.animate.set_fill(BLUE, opacity=0.5) for neuron in output_layer],
            run_time=1
        )
```
